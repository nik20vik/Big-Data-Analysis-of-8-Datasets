### Q. 3) In which Region, maximum number of Confirmed cases were recorded ?

```python
data.groupby('Region')['Confirmed'].sum().sort_values(ascending = False)
```

### Q. 4) In which Region, minimum number of Deaths cases were recorded ?

### Q. 6-A ) Sort the entire data wrt No. of Confirmed cases in ascending order.

```python
# by default is ascending but can use 'True' here too
# data.sort_values(by = ['Confirmed'])
data.sort_values(by = ['Confirmed'], ascending = True).head(20)
```

```python
# Sorting using more columns
data.sort_values(by = ['Confirmed', 'Deaths'], ascending = True).head(20)
```

### Q. 6-B ) Sort the entire data wrt No. of Recovered cases in descending order.
